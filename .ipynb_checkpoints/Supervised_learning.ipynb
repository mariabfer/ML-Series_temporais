### Start of the Pipeline
pulls in of data

```python
from sklearn.datasets import fetch_california_housing
#dataset das casas da california, tem 8 atributos e 20640 amostras
housing = fetch_california_housing(as_frame=True)

house = housing.frame

house.head()
```

### visualize the data

```python

import seaborn as sns
import pandas as pd

house["price_house"] = pd.cut(house["MedHouseVal"],
                           bins=[0,2,4,6],
                           labels=["baixo","medio","alto"])

sns.pairplot(
    house,
    vars=["MedInc", "HouseAge", "AveRooms", "MedHouseVal", "Population"],
    hue="price_house",
    height=3
)
```

### remoçao da variavel 'MedHouseVal', a fim dela ficar a função das outras variaveis.

```python
X_house = house.drop('MedHouseVal', axis=1) #usar todas as variaiveis, exceto 'MedHouseVal'
print('X_house.shape:', X_house.shape)
y_house = house['MedHouseVal'] #usar somente a variavel 'MedHouseVal' como resposta  
print('y_house.shape:', y_house.shape)
```
